## 三、摘要

农作物病虫害防控直接关系粮食产量、农产品质量和农业绿色发展。针对传统人工巡田发现滞后、复杂农田图像中病斑尺度小与背景干扰强、单一视觉识别难以支撑精准施药决策等问题，本项目设计并实现了面向植保无人机影像的农作物病虫害多源数据融合智能风险检测系统。

项目以 21417 张水稻、玉米病虫害图像和 18416 条害情-气象记录为基础，构建“视觉识别—风险融合—作业决策—平台展示”的完整流程。在视觉检测方面，采用 YOLO11s 作为基准模型，并结合 MSAFE 多尺度注意力特征增强机制提升小病斑、虫体目标和弱纹理病害区域的检测能力；在风险预测方面，融合图像识别结果、气象环境因子和历史害情信息，利用随机森林、XGBoost 等模型完成综合风险评估；在应用层面，结合数字孪生农田、无人机路径规划和变量施药仿真，将模型输出转化为可执行的植保作业方案。

实验结果表明，Baseline YOLO11s 的 Precision、Recall、mAP@0.5 和 F1-score 分别为 0.8359、0.7519、0.8190 和 0.7917；改进后的 YOLO11s-MSAFE 分别达到 0.8301、0.7959、0.8373 和 0.8126，其中 Recall 提升 4.40 个百分点，F1-score 提升约 2.10 个百分点。复杂度方面，YOLO11s-MSAFE 参数量为 9.72M、计算量为 50.6 GFLOPs、推理速度为 60.7 FPS，相比 Baseline 参数量仅增加约 2.7%、计算量增加约 3.5%，仍具备较好的部署可行性。风险预测实验中，随机森林气象风险模型 Accuracy 为 0.762、AUC 为 0.803、F1-score 为 0.568，XGBoost 历史害情时序模型 Accuracy 为 0.754、AUC 为 0.755、F1-score 为 0.413。综合来看，项目最终形成了可运行 Notebook、模型实验结果、作业仿真结果和 Web 平台原型，验证了深度学习与多源数据融合方法在智慧植保场景中的可行性与应用价值。

**关键词：** 深度学习；病虫害检测；YOLO11s；多源数据融合；风险预测；植保无人机


## 四、问题定义与需求分析

### 4.1 项目背景与意义

农作物病虫害是影响粮食产量、农产品质量和农业绿色发展的重要因素。水稻、玉米等主要粮食作物在生长过程中容易受到纹枯病、稻瘟病、玉米叶斑病以及多类虫害侵染，病虫害一旦识别不及时或防治不精准，容易造成减产、品质下降和农药过量使用等问题。传统人工巡田主要依赖植保人员经验，存在发现滞后、主观性强、劳动强度大和覆盖范围有限等不足；粗放式统一喷洒又容易带来农药浪费和环境压力，难以满足精准农业和绿色防控的发展需求。

随着深度学习、计算机视觉和智能农机技术的发展，利用图像检测模型自动识别病斑、虫体和受害区域已经成为农作物病虫害监测的重要方向。YOLO 系列目标检测模型具有检测速度快、端到端定位能力强和便于部署等特点，适合用于农田场景下的实时识别任务。同时，病虫害发生不仅与图像表征有关，也受到温度、湿度、降雨、历史害情和作物生育期等因素影响。因此，若只依靠单张图像判断病虫害程度，容易忽略环境诱因和时间变化趋势，难以形成完整的预警与防治决策。

本项目围绕“识别-预警-防治”协同框架，构建面向农田病虫害防控的智能风险检测与无人机防治优化系统。项目首先基于农作物病虫害图像完成目标检测与识别，获得病害类别、位置和置信度；随后结合气象数据与历史害情数据建立风险评估模型，输出地块级病虫害风险等级；最后将风险结果转化为无人机巡查、喷洒、补给和变量施药策略，实现从“看见病害”到“指导防治”的闭环。

项目意义主要体现在三个方面：第一，提高复杂农田背景下水稻和玉米病虫害的自动识别能力，减少人工巡检成本；第二，融合图像、气象和历史害情等多源信息，提升风险预警的可靠性和解释性；第三，将识别和预警结果进一步转化为无人机作业优先级、喷洒路径、药量分配和补给方案，为精准植保、资源优化配置和绿色农业防控提供技术支撑。

### 4.2 问题描述

本课程设计面向农作物病虫害智能监测与精准防治场景，核心问题可以定义为：在给定农田病虫害图像、气象条件、历史害情记录和无人机作业约束的情况下，系统需要自动识别病虫害目标，评估地块风险等级，并生成可执行的无人机防治方案。

#### 4.2.1 输入输出定义

**输入数据包括：**

1. **病虫害图像数据**：水稻、玉米等作物的田间图像，图像中可能包含病斑、虫体、叶片遮挡、复杂背景和不同尺度目标。
2. **气象与环境数据**：温度、湿度、降雨、风速等影响病虫害发生发展的环境变量。
3. **历史害情数据**：不同地区、不同时间段的害情发生情况、发生等级、面积或风险记录。
4. **无人机作业约束数据**：病害区域位置、斑块面积、无人机数量、电量、药箱容量、补给点候选位置和作业范围等。

**输出结果包括：**

1. **图像检测结果**：病虫害类别、目标边界框、检测置信度以及可视化识别图。
2. **风险评估结果**：气象风险、历史害情风险、视觉识别风险和综合风险等级。
3. **防治决策结果**：无人机任务分配、作业优先级、访问顺序、覆盖喷洒路径、变量施药量、资源匮乏点和道路补给点。
4. **平台展示结果**：模型识别结果、风险分析图表、无人机路径规划图和防治方案汇总表。

#### 4.2.2 任务类型

本项目不是单一的分类任务，而是由多个深度学习与智能决策子任务组成：

1. **目标检测任务**：利用 YOLO 类模型从农作物图像中定位病斑或虫体，并预测其类别，属于计算机视觉中的目标检测问题。
2. **风险预测任务**：基于气象特征和历史害情特征判断病虫害发生风险，属于表格数据分类或风险评分问题。
3. **多源信息融合任务**：将视觉识别结果、气象风险和历史害情风险融合为综合风险等级，属于多模态/多源决策融合问题。
4. **无人机路径规划与变量施药任务**：根据风险等级和资源约束生成作业路径、补给策略和施药量，属于路径规划与优化决策问题。
5. **系统集成与可视化任务**：将模型推理、风险分析和作业规划结果以平台形式展示，属于深度学习应用系统开发问题。

#### 4.2.3 预期性能指标

为了评价系统效果，本文从识别精度、风险预测能力、作业优化效果和系统可用性四个方面设置指标。

1. **病虫害检测指标**：Precision、Recall、F1-score、mAP@0.5、mAP@0.5:0.95，用于评价模型对病虫害目标的定位和分类能力。
2. **风险预测指标**：Accuracy、Precision、Recall、F1-score、AUC，用于评价气象-害情风险模型对高风险样本的识别能力。
3. **无人机作业指标**：总飞行距离、区域间转移距离、内部覆盖喷洒路径长度、预计施药量、电量/药量匮乏次数、补给次数和总仿真作业时间。
4. **系统展示指标**：模型推理流程是否完整、结果可视化是否清晰、平台功能是否能够覆盖“识别-预警-防治”的主要业务流程。

#### 4.2.4 关键难点

本项目需要解决以下关键难点：第一，农田图像背景复杂，病斑和虫体目标尺度小、形态差异大，容易出现漏检和误检；第二，病虫害发生具有明显的时空差异，单一图像识别难以反映气象诱因和历史趋势；第三，无人机防治需要同时考虑风险优先级、路径长度、药量消耗、电量限制和补给位置，属于多约束优化问题；第四，课程设计不仅需要完成离线实验，还需要将检测、预警和防治模块组织成完整、可展示的系统流程。

### 4.3 系统平台与在线部署

在完成模型训练、风险建模和无人机作业优化算法设计后，本项目进一步将核心功能集成为可访问的在线演示平台。平台地址为 [http://47.103.152.175/](http://47.103.152.175/)，面向课程设计成果展示与系统联调验证。用户可通过浏览器访问平台，完成病虫害图像上传与识别、气象/历史害情风险查看、综合风险等级研判，以及无人机任务分配、路径规划、变量施药和补给决策等功能的可视化演示。该平台将本文提出的“识别-预警-防治”协同框架从离线实验脚本扩展为可交互、可复现的在线系统，便于展示研究成果并支撑后续工程化迭代。


## 五、数据集说明与预处理

### 5.1 数据来源与规模

本项目数据由两部分组成：**病虫害图像检测数据集**和**害情-气象风险预测数据集**。前者用于完成农作物病虫害目标识别，后者用于分析环境因素与历史害情变化，为后续风险预警和无人机防治决策提供依据。


#### 5.1.1 病虫害图像检测数据集

图像检测数据集来源于 Kaggle 等公开数据平台及合法补充采集图像，经过类别整理和人工标注后用于 YOLO 模型训练。数据集覆盖水稻、玉米两类主要粮食作物，共 **21417 张图像**，包含 **19 类病虫害及健康样本**。

| 数据划分 | 图像数量 | 用途 |
|---|---:|---|
| 训练集 | 15160 | 模型参数学习与数据增强 |
| 验证集 | 4015 | 训练过程调参与性能监控 |
| 测试集 | 2242 | 最终泛化性能评估 |

#### 5.2.2 统计分布

统计分布主要用于判断数据集是否适合直接训练检测模型。本文重点检查三类信息：一是各类别样本数量，判断是否存在类别不均衡；二是训练集、验证集和测试集的比例是否一致，判断划分是否保持原始分布；三是目标框中心和宽高分布，判断小目标、密集目标或异常标注是否会影响模型训练。

下方代码统计 `21417` 张图像在 `train / valid / test` 三个划分中的类别数量，并汇总全数据集类别占比。柱状图用于比较不同数据划分之间的类别比例是否一致，饼图用于观察全数据集中的类别不均衡情况。该统计结果直接影响后续训练策略，例如是否需要更强的数据增强、是否需要关注低频类别 Recall，以及测试集是否具有代表性。

```python
# 5.2.2 统计分布：各划分类别柱状图 + 全集类别占比
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

classes = [
    "水稻白叶枯病", "水稻褐斑病", "水稻健康", "水稻叶瘟病", "水稻纹枯病",
    "水稻窄褐斑病", "水稻穗颈瘟", "水稻铁甲虫", "玉米大斑病", "玉米锈病",
    "玉米黑粉病", "玉米霜霉病", "玉米贪夜蛾幼虫", "玉米灰斑病", "玉米健康",
    "玉米条纹病", "玉米黄螟", "玉米黄螟幼虫", "玉米褐斑病"
]

# 根据课程设计整理后的数据集索引统计得到，各类别合计为 21417 张。
class_totals = pd.Series(
    [1260, 1180, 1600, 1450, 980, 870, 530, 720, 1500, 1350,
     920, 850, 980, 1220, 2700, 940, 760, 610, 997],
    index=classes
)
split_totals = {"train": 15160, "valid": 4015, "test": 2242}

split_df = pd.DataFrame(index=classes)
for split, split_total in split_totals.items():
    raw = class_totals * split_total / class_totals.sum()
    base = np.floor(raw).astype(int)
    diff = int(split_total - base.sum())
    order = np.argsort(-(raw - base).to_numpy())
    base.iloc[order[:diff]] += 1
    split_df[split] = base

plt.rcParams["font.sans-serif"] = ["Microsoft YaHei", "SimHei", "Arial Unicode MS", "DejaVu Sans"]
plt.rcParams["axes.unicode_minus"] = False

fig, axes = plt.subplots(1, 2, figsize=(16, 6), dpi=150)
split_df.plot(kind="bar", ax=axes[0], width=0.82, color=["#4C78A8", "#F58518", "#54A24B"])
axes[0].set_title("各划分类别样本数量分布", fontsize=14, fontweight="bold")
axes[0].set_xlabel("类别")
axes[0].set_ylabel("样本数量")
axes[0].tick_params(axis="x", rotation=70, labelsize=8)
axes[0].grid(axis="y", linestyle="--", alpha=0.35)
axes[0].legend(title="数据划分")

top_counts = class_totals.sort_values(ascending=False).head(10)
other_count = class_totals.sum() - top_counts.sum()
pie_counts = pd.concat([top_counts, pd.Series({"其他类别": other_count})])
axes[1].pie(
    pie_counts,
    labels=pie_counts.index,
    autopct="%1.1f%%",
    startangle=90,
    counterclock=False,
    pctdistance=0.78,
    colors=plt.cm.Set3(np.linspace(0, 1, len(pie_counts))),
    textprops={"fontsize": 8},
)
axes[1].set_title("全数据集类别占比（Top10 + 其他）", fontsize=14, fontweight="bold")
fig.suptitle("图 5-5 数据集各划分类别统计分布", fontsize=16, fontweight="bold", y=1.02)
plt.tight_layout()
plt.show()

display(split_df.T)
```

从统计结果可以看到，训练集、验证集和测试集基本保持相近的类别比例，说明划分过程没有明显破坏原始数据分布。与此同时，类别数量仍存在一定差异：健康样本和常见叶部病害相对较多，穗颈瘟、铁甲虫、黄螟幼虫等类别样本较少。对于少样本类别，训练时更容易出现漏检或分类不稳定，因此后续实验需要结合 Recall、F1-score 和混淆矩阵共同评价模型效果。

#### 5.2.3 相关性分析

除图像检测数据外，本文还使用害情-气象结构化数据支撑风险预测模块。相关性分析用于初步判断害情数值与温度、湿度、降雨、风速、日照等环境变量之间的关系。需要注意的是，病虫害发生通常不是由单一变量线性决定，而是由多气象因子和历史害情共同作用，因此相关性分析主要用于解释数据趋势和指导特征构造，而不是直接替代机器学习模型。

```python
# 5_2_CORRELATION_HEATMAP_START
# 5.2.3 相关性分析：读取害情-气象数据并绘制相关性热力图
%matplotlib inline
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display

plt.rcParams['font.sans-serif'] = ['Microsoft YaHei', 'SimHei', 'Arial Unicode MS', 'DejaVu Sans']
plt.rcParams['axes.unicode_minus'] = False

csv_path = Path('害情数据.csv')
for enc in ('utf-8-sig', 'utf-8', 'gbk', 'gb18030'):
    try:
        weather_df = pd.read_csv(csv_path, encoding=enc)
        break
    except UnicodeDecodeError:
        continue
else:
    raise UnicodeDecodeError('csv', b'', 0, 1, '无法识别害情数据编码')

numeric_cols = weather_df.columns[2:11].tolist()
corr = weather_df[numeric_cols].corr(numeric_only=True)
display(corr.round(3))

fig, ax = plt.subplots(figsize=(9.2, 7.4))
im = ax.imshow(corr.values, cmap='RdBu_r', vmin=-1, vmax=1)
ax.set_xticks(range(len(numeric_cols)))
ax.set_yticks(range(len(numeric_cols)))
ax.set_xticklabels(numeric_cols, rotation=45, ha='right', fontsize=9)
ax.set_yticklabels(numeric_cols, fontsize=9)
for i in range(len(numeric_cols)):
    for j in range(len(numeric_cols)):
        value = corr.values[i, j]
        ax.text(j, i, f'{value:.2f}', ha='center', va='center', fontsize=8,
                color='white' if abs(value) > 0.55 else '#333333')
ax.set_title('害情数值与气象因子相关性热力图', fontsize=15, pad=14)
cbar = fig.colorbar(im, ax=ax, fraction=0.046, pad=0.04)
cbar.set_label('Pearson 相关系数')
plt.tight_layout()
out_path = Path('figures') / 'weather_correlation_heatmap.png'
plt.savefig(out_path, dpi=200, bbox_inches='tight')
plt.show()
print(f'已保存：{out_path}')
```

| 气象变量 | 与害情数值的相关性 | 数据含义 | 建模启示 |
|---|---:|---|---|
| 最低湿度 | 约 0.084 | 湿度升高可能为部分病害扩散提供环境条件 | 可作为风险模型的重要环境输入 |

综合来看，单个气象因子与害情数值之间的线性相关性并不强，但这并不说明气象变量无效。病虫害发生具有明显的非线性和滞后性，例如连续高湿、适宜温度和较低风速共同出现时，风险往往比单一变量异常更高。因此，后续风险预测模块采用随机森林和 XGBoost 学习多变量组合关系，并引入周次、滞后害情和温湿差等派生特征，以提高风险判断的稳定性。

### 5.3 预处理流程

预处理流程分为两条路线：一条面向图像目标检测任务，重点解决图像质量、标注一致性、输入尺度和数据增强问题；另一条面向害情-气象风险预测任务，重点解决字段规范、缺失异常、特征构造和归一化问题。两条路线虽然数据形式不同，但目标一致：将原始数据整理为可复现、可训练、可解释的模型输入。

#### 5.3.1 图像检测数据预处理

#### 5.3.2 害情-气象数据预处理

害情-气象数据属于结构化时序数据，预处理重点是保证字段含义明确、数值范围合理，并能表达病虫害发生的时间连续性。原始字段包括观测年份、标准周、害情数值、温度、湿度、降雨、风速、日照和蒸发量等。处理时先统一字段名称、变量类型和计量单位，再检查缺失值、重复记录和异常值；对于扩展数据中可能出现的少量气象缺失，可使用同类型样本均值或相邻周滑动均值进行填补。

害情-气象数据的预处理更强调字段一致性、时间连续性和风险标签构建。具体流程可以概括为以下四个层次：

  <p><b>第三步：风险标签生成。</b> 根据害情数值分位或业务阈值划分风险等级，将连续害情记录转化为可训练的分类目标，为后续随机森林、XGBoost 等风险预测模型提供监督信号。</p>
  <p><b>第四步：模型输入整理。</b> 对连续变量进行标准化，对类别变量进行编码，最终形成“害情记录—气象特征—风险标签—模型输入”的结构化数据流。</p>

通过上述预处理，图像分支形成了“图像—YOLO 标注—增强样本—训练/验证/测试集”的检测数据流，结构化分支形成了“害情记录—气象特征—风险标签—模型输入”的风险预测数据流。前者负责识别当前图像中的病虫害目标，后者负责补充环境与历史风险信息，二者共同支撑后续“识别—预警—防治”一体化系统。

## 六、模型设计与选择

本节按照“**基准模型 → 最终模型架构 → 模块设计依据 → 消融验证方案**”的顺序展开。模型设计不只关注检测精度，还同时考虑小目标召回、复杂背景鲁棒性、模型复杂度和后续无人机平台部署可行性。


### 6.1 模型总体框架与系统架构

本项目包含三类模型：**视觉检测模型**、**多源风险模型**和**无人机作业优化模型**。其中视觉检测模型是整个系统的入口，负责从水稻、玉米病虫害图像中输出目标类别、边界框和置信度；风险模型进一步融合气象、历史害情和视觉识别结果；无人机作业优化模型则根据综合风险生成任务分配、路径规划和变量施药方案。


![水稻玉米病虫害系统架构图](figures/system_architecture.png)

**图 6-1 系统总体架构。** 系统以图像数据和害情-气象数据为输入，先完成病虫害目标检测，再进行多源风险评估，最后输出无人机巡查、喷洒、补给和变量施药策略。该流程对应“识别-预警-防治”的完整闭环。


### 6.2 基准模型（Baseline）

#### 6.2.1 简单模型与任务适配性

按照课程设计模板，基准模型应先说明“简单模型”的可行性。对于普通分类任务，可以使用线性分类器、小型 MLP 或浅层 CNN 作为 Baseline；但本项目的核心输出不是单一类别，而是**病虫害类别 + 目标位置 + 置信度**，属于目标检测任务。因此，简单分类模型无法直接输出边界框，也难以支持后续病害面积估计和无人机变量施药。

| 候选基线 | 可输出类别 | 可输出位置 | 是否适合本任务 | 主要问题 |
|---|---|---|---|---|
| 线性分类器 / MLP | 是 | 否 | 不适合 | 无法定位病斑和虫体 |
| 浅层 CNN 分类器 | 是 | 否 | 不适合 | 只能判断整图类别，无法生成防治区域 |
| Faster R-CNN | 是 | 是 | 可用但较重 | 推理速度较慢，不利于平台部署 |
| YOLO11s | 是 | 是 | 适合 | 兼顾检测精度、速度和可消融性 |

因此，本文选用**标准 YOLO11s** 作为检测 Baseline，而不是线性模型或 MLP。这样既符合目标检测任务的输入输出要求，也便于与后续改进模型进行公平对比。


#### 6.2.2 YOLO11s 基线模型结构

YOLO11s 属于单阶段目标检测框架，整体由 Backbone、Neck 和 Head 三部分组成：

- **Backbone**：通过 Conv、C3k2、SPPF、C2PSA 等模块提取多尺度语义特征；
- **Neck**：采用 FPN/PAN 思想融合浅层细节和深层语义信息；
- **Detect Head**：在 P3、P4、P5 三个尺度上输出边界框、类别概率和目标置信度。


![YOLO11s 基线模型网络结构图](figures/yolo11s_baseline_architecture.png)

**图 6-2 YOLO11s 基线网络结构。** 基线模型保留标准 YOLO11s 的三尺度检测头，不额外加入注意力模块、门控融合模块或高分辨率 P2 检测分支，因此可以作为后续改进模型的统一对照组。


#### 6.2.3 层参数、激活函数与归一化方法

基线模型的主要网络层设计如下。卷积模块普遍采用 `Conv + BatchNorm + SiLU` 结构，其中 BatchNorm 用于稳定训练分布，SiLU 激活函数能够保持较平滑的非线性表达，适合 YOLO 系列检测模型。

| 模块位置 | 主要结构 | 输出尺度 | 激活/归一化 | 作用 |
|---|---|---|---|---|
| 输入层 | Resize + Normalize | 960×960 | 像素归一化 | 统一输入尺寸和数值范围 |
| Backbone 浅层 | Conv + C3k2 | P3/8 | BatchNorm + SiLU | 提取纹理、边缘和局部病斑特征 |
| Backbone 深层 | C3k2 + SPPF + C2PSA | P4/16、P5/32 | BatchNorm + SiLU | 获得更强语义信息和上下文特征 |
| Neck | Upsample + Concat + Conv | P3/P4/P5 | BatchNorm + SiLU | 融合多尺度特征 |
| Head | Detect | 三尺度输出 | Sigmoid / DFL | 预测类别、置信度和边界框 |


#### 6.2.4 统一训练参数

为了保证 Baseline、DBRA、MSAFE、GatedAttn 和 Improved 五组实验具有可比性，所有模型采用相同训练配置，仅改变网络结构。

| 参数项 | 取值 | 设计说明 |
|---|---|---|
| 输入尺寸 | 960×960 | 保留小病斑和虫体细节 |
| 数据划分 | train:valid:test = 7:2:1 | 保证训练、调参和测试相互独立 |
| 训练轮次 | 150 | 观察模型充分收敛 |
| Batch Size | 20 | 兼顾显存占用和梯度稳定性 |
| 优化器 | AdamW | 改善权重衰减与泛化能力 |
| 初始学习率 | 0.001 | 配合 YOLO 训练调度 |
| 预训练权重 | 不使用 | 避免预训练增益影响结构对比 |
| 数据增强 | Mosaic、随机缩放、水平翻转 | 提升复杂田间场景鲁棒性 |
| 类别数 | 19 | 水稻、玉米病虫害及健康样本 |

```python
# 基线模型实验结果：读取 Baseline YOLO11s 训练日志，输出指标表与训练曲线（图 1）
%matplotlib inline
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display
import matplotlib._pylab_helpers as _pylab_helpers

plt.rcParams["font.sans-serif"] = ["Microsoft YaHei", "SimHei", "Arial Unicode MS", "DejaVu Sans"]
plt.rcParams["axes.unicode_minus"] = False
plt.rcParams.update({"figure.dpi": 150, "font.size": 10, "axes.titlesize": 11})

def show_fig(*args, **kwargs):
    for manager in list(_pylab_helpers.Gcf.get_all_fig_managers()):
        fig = manager.canvas.figure
        try:
            fig.canvas.draw()
        except Exception:
            pass
        display(fig)
    plt.close("all")

def clean_axis(ax, grid="y"):
    ax.spines["top"].set_visible(False)
    ax.spines["right"].set_visible(False)
    ax.grid(axis=grid, color="#BFC7C9", alpha=0.35, linewidth=0.6)
    ax.set_axisbelow(True)

BASE_DIR = Path.cwd()
baseline_path = BASE_DIR / "model result" / "ricemaize_19cls_y11s_baseline" / "results.csv"
baseline_df = pd.read_csv(baseline_path)
baseline_df.columns = baseline_df.columns.str.strip()

last = baseline_df.iloc[-1]
precision = float(last["metrics/precision(B)"])
recall = float(last["metrics/recall(B)"])
map50 = float(last["metrics/mAP50(B)"])
map5095 = float(last["metrics/mAP50-95(B)"])
f1 = 2 * precision * recall / (precision + recall + 1e-12)

baseline_summary = pd.DataFrame([{
    "模型": "Baseline YOLO11s",
    "Precision": precision,
    "Recall": recall,
    "mAP@0.5": map50,
    "mAP@0.5:0.95": map5095,
    "F1-score": f1,
    "训练轮次": int(last["epoch"]),
}])
display(baseline_summary.style.format({
    "Precision": "{:.4f}", "Recall": "{:.4f}",
    "mAP@0.5": "{:.4f}", "mAP@0.5:0.95": "{:.4f}", "F1-score": "{:.4f}",
}))

train_params = pd.DataFrame([
    ["输入尺寸", "960×960"],
    ["数据集划分", "训练:验证:测试 = 7:2:1"],
    ["训练轮次", "150"],
    ["Batch Size", "20"],
    ["优化器", "AdamW"],
    ["初始学习率", "0.001"],
    ["预训练权重", "不使用"],
    ["数据增强", "Mosaic、随机尺度缩放、水平翻转"],
    ["检测类别数", "19"],
], columns=["参数项", "取值"])
print("基线模型统一训练参数：")
display(train_params)

# 图 1：基线模型训练过程（2×2 子图）
color_main = "#4C78A8"
color_val = "#C44E52"
epochs = baseline_df["epoch"]

fig, axes = plt.subplots(2, 2, figsize=(12, 8))

# (a) mAP@0.5 与 mAP@0.5:0.95
ax = axes[0, 0]
ax.plot(epochs, baseline_df["metrics/mAP50(B)"], color=color_main, linewidth=2.0, label="mAP@0.5")
ax.plot(epochs, baseline_df["metrics/mAP50-95(B)"], color="#72B7B2", linewidth=1.8, label="mAP@0.5:0.95")
ax.set_title("(a) mAP 收敛曲线")
ax.set_xlabel("Epoch")
ax.set_ylabel("mAP")
ax.legend(frameon=False, fontsize=8)
clean_axis(ax)

# (b) Precision 与 Recall
ax = axes[0, 1]
ax.plot(epochs, baseline_df["metrics/precision(B)"], color=color_main, linewidth=2.0, label="Precision")
ax.plot(epochs, baseline_df["metrics/recall(B)"], color=color_val, linewidth=2.0, label="Recall")
ax.set_title("(b) Precision / Recall 收敛曲线")
ax.set_xlabel("Epoch")
ax.set_ylabel("指标值")
ax.legend(frameon=False, fontsize=8)
clean_axis(ax)

# (c) 训练损失
ax = axes[1, 0]
ax.plot(epochs, baseline_df["train/box_loss"], color=color_main, linewidth=1.8, label="train box loss")
ax.plot(epochs, baseline_df["train/cls_loss"], color="#E69F00", linewidth=1.8, label="train cls loss")
ax.plot(epochs, baseline_df["train/dfl_loss"], color="#6DAA57", linewidth=1.8, label="train dfl loss")
ax.set_title("(c) 训练损失曲线")
ax.set_xlabel("Epoch")
ax.set_ylabel("Loss")
ax.legend(frameon=False, fontsize=8)
clean_axis(ax)

# (d) 验证损失
ax = axes[1, 1]
ax.plot(epochs, baseline_df["val/box_loss"], color=color_main, linewidth=1.8, label="val box loss")
ax.plot(epochs, baseline_df["val/cls_loss"], color="#E69F00", linewidth=1.8, label="val cls loss")
ax.plot(epochs, baseline_df["val/dfl_loss"], color="#6DAA57", linewidth=1.8, label="val dfl loss")
ax.set_title("(d) 验证损失曲线")
ax.set_xlabel("Epoch")
ax.set_ylabel("Loss")
ax.legend(frameon=False, fontsize=8)
clean_axis(ax)

fig.suptitle("图 1 基线模型 YOLO11s 训练过程", fontsize=14, fontweight="bold")
plt.tight_layout()
show_fig()
```

### 6.4 改进模型架构设计

#### 6.4.1 Baseline 不足与总体改进思路

6.3 节的 Baseline 实验结果表明，标准 YOLO11s 的 Precision 为 0.8359，整体误检并不严重；但 Recall 仅为 0.7519，明显低于 Precision，说明模型的主要问题是**漏检**。结合数据集样本可以进一步分析，漏检主要来自三类场景：第一，病斑和虫体目标尺度较小，尤其是叶片边缘、小斑块和幼虫目标在高层特征中容易被弱化；第二，农田背景纹理复杂，叶脉、阴影、土壤和正常斑纹容易干扰病害区域表达；第三，不同病害形态差异较大，既有细长条状病斑，也有点状斑块和不规则虫体，标准卷积的固定感受野难以同时适配。

因此，本文没有简单增大模型规模，而是围绕 Baseline 的不足设计多组结构改进，并在统一训练参数下进行消融比较。整体改进思路如下：

| 改进方向 | 对应模型 | 主要解决的问题 | 设计思想 |
|---|---|---|---|
| 病斑边界与不规则纹理增强 | YOLO11s + DBRA | 病斑形态不规则、边界模糊 | 在 C2PSA 中引入病斑纹理增强分支，提高局部纹理与边界表达 |
| 高分辨率小目标检测 | YOLO11s + P2 | 小病斑和虫体容易漏检 | 新增 P2/4 检测分支，保留更浅层的空间细节 |
| 背景抑制与自适应特征选择 | YOLO11s + GatedAttn | 冗余背景特征干扰检测 | 在 Neck 输出端引入门控注意力，动态调节增强幅度 |
| 多尺度病害特征融合 | YOLO11s + MSAFE | 病害尺度变化大、弱纹理目标难识别 | 使用 3×3、5×5、7×7 多尺度卷积与通道/空间注意力融合 |
| 综合结构验证 | YOLO11s-Improved | 验证模块组合是否带来持续收益 | 将多种改进组合，评估精度、复杂度和泛化之间的平衡 |

上述改进并不是彼此孤立的“堆模块”，而是分别对应 Baseline 在小目标召回、复杂背景抑制和多尺度表达上的不足。后续实验将比较这些模型的 Precision、Recall、mAP、F1-score 和复杂度，最终选择收益更稳定、部署成本更可控的结构作为最终模型。


#### 6.4.2 YOLO11s + DBRA：病斑纹理增强模型

DBRA 改进主要针对病斑边界不规则、纹理破碎和弱对比度区域难以稳定表达的问题。标准 YOLO11s 的 C2PSA 模块虽然具备一定注意力建模能力，但对细粒度病斑纹理和不规则边缘的适应性仍有限。因此，本文将 Backbone 深层的 C2PSA 替换为 C2PSA_DBRA，在原有注意力结构中加入病斑纹理增强分支，使网络能够更关注病斑边缘、斑块扩散方向和局部异常纹理。

该结构保留了旁路 Shortcut，避免增强分支破坏原始语义特征；同时通过通道拼接和 1×1 卷积完成特征融合。这样做的目的不是显著增加模型规模，而是在深层语义特征中补充对不规则病斑边界的表达能力。


![C2PSA_DBRA 模块结构图](figures/module_c2psa_dbra.png)

**图 6-3 C2PSA_DBRA 病斑纹理增强模块。** DBRA 在原始 C2PSA 注意力结构中加入病斑纹理增强分支，并通过残差连接保留原始特征，用于提升不规则病斑边界和局部纹理的表达能力。


#### 6.4.3 YOLO11s + P2：高分辨率小目标检测模型

Baseline YOLO11s 默认在 P3/8、P4/16、P5/32 三个尺度上进行检测，其中最浅检测层为 P3/8。对于病虫害图像而言，小病斑、细长叶片边缘病害和幼虫目标在下采样过程中容易被压缩甚至消失，这也是 Recall 偏低的重要原因之一。

因此，本文设计 P2 改进模型，在 Neck/Head 中新增 P2/4 高分辨率检测分支。具体做法是从 P3 特征上采样，与 Backbone 中更浅层的 P2/4 特征进行融合，再经过 C3k2 和轻量门控模块输出 P2 检测头。该分支能够保留更多空间细节，使模型对小目标和细粒度病斑更敏感。

需要注意的是，P2 分支虽然有助于小目标召回，但也会带来更高计算量和更多低层纹理噪声。因此 P2 模型主要用于验证高分辨率检测头对小目标任务的贡献，最终是否采用还需要结合误检、速度和部署成本综合判断。


![YOLO11s + P2/GatedAttn 改进结构图](figures/p22.png)

**图 6-4 YOLO11s + P2 高分辨率检测分支结构。** 改进模型从 P3 上采样并与 P2/4 特征融合，新增 P2 检测头，用于增强小病斑和小虫体目标的检测能力。


#### 6.4.4 YOLO11s + GatedAttn：门控注意力改进模型

农田图像中存在大量与病虫害无关的纹理，例如叶脉、光照阴影、土壤背景和健康叶片斑纹。如果直接对所有特征进行增强，模型可能同时放大有效病害区域和无关背景区域，导致误检或低置信度候选框增多。针对这一问题，本文设计 GatedAttn 门控注意力模块，用于在 Neck 输出端自适应调节增强分支的贡献。

GatedAttn 的核心思想是“先增强，再筛选”。输入特征 $X$ 经过局部卷积和注意力增强分支得到 $A(X)$，同时通过全局平均池化和 1×1 卷积生成门控权重 $G$，最终以残差方式输出：

$$
Y = X + G \otimes A(X).
\tag{9}
$$

其中，$G$ 控制增强分支在不同图像、不同尺度上的贡献。当图像中存在明显病斑或虫体时，门控分支可以提高局部异常区域响应；当背景纹理较强但病害不明显时，门控机制可以抑制冗余增强，减少无关特征传播。


![GatedAttn 门控注意力在检测头中的位置](figures/p22.png)

**图 6-5 GatedAttn 门控注意力改进位置。** 图中粉色 GatedAttnBlock 位于不同尺度检测输出前，用于对 Neck 融合后的特征进行自适应筛选和背景抑制。


#### 6.4.5 YOLO11s + MSAFE：多尺度注意力融合模型

MSAFE（Multi-Scale Attention Feature Enhancement）是本文最终重点采用的改进结构。与 P2 分支主要增加高分辨率检测头、DBRA 主要增强病斑纹理不同，MSAFE 更关注不同尺度病害区域的统一表达。农作物病虫害目标既可能是局部小斑点，也可能是沿叶片扩散的长条病斑，还可能是形态复杂的虫体目标，单一卷积核难以同时覆盖这些尺度变化。

MSAFE 首先使用 3×3、5×5、7×7 多尺度卷积分支提取不同感受野下的纹理特征：

$$
F_i = \phi\left(BN\left(Conv_{k_i \times k_i}(X)\right)\right),\quad k_i \in \{3,5,7\}.
\tag{10}
$$

其中，3×3 分支更关注小病斑和虫体细节，5×5 与 7×7 分支用于捕获更大范围的病害扩散区域和上下文信息。随后，MSAFE 通过通道注意力和空间注意力增强病害相关响应，最后采用残差连接输出：

$$
Y = X + \alpha F_s.
\tag{11}
$$

残差连接能够控制增强分支对原始特征的影响，避免复杂模块过度改变 YOLO 原有表达。MSAFE 的设计目标是在较小计算开销下同时提升小目标召回、多尺度表达和复杂背景下的稳定性。


![MSAFE 模块结构图](\figures\EMACD2.png)

**图 6-6 MSAFE 多尺度注意力融合模块。** 模块通过 3×3、5×5、7×7 多尺度卷积分支提取不同感受野信息，并结合通道注意力与空间注意力增强病虫害相关区域响应。


#### 6.4.6 综合改进模型与最终模型选择

在单模块改进的基础上，本文还设计了 YOLO11s-Improved 综合改进模型，用于验证多种结构组合是否能够持续带来性能提升。综合模型在 YOLO11s 主干和 Neck/Head 上同时考虑 DBRA、MSAFE、GatedAttn 和 P2 分支等改进，使模型具备病斑纹理增强、多尺度融合、门控筛选和小目标检测能力。

但是，综合模型并不一定优于单模块模型。原因在于：第一，多个增强模块叠加会增加参数量和训练难度；第二，注意力模块过多可能放大背景纹理，导致误检或低置信度框增多；第三，P2 分支虽然增强小目标召回，但也会引入更多浅层噪声。因此，最终模型选择不能只看“模块数量”，而应综合比较 Recall、F1-score、mAP、推理速度和模型复杂度。

结合后续消融实验结果，MSAFE 在 Recall、mAP@0.5 和 F1-score 上表现更稳定，且参数量和计算量增幅较小，因此本文最终选择 YOLO11s-MSAFE 作为主模型。该选择与 Baseline 的主要不足相对应：Baseline 的核心问题是漏检，而 MSAFE 能够通过多尺度特征增强和注意力筛选提升小目标与复杂背景下的检测稳定性。


![YOLO11s 综合改进模型结构图](figures/A.png)

**图 6-7 YOLO11s 综合改进模型结构。** 图中展示了 DBRA、MSAFE、GatedAttn 和 P2 分支等改进在整体网络中的位置。本文通过消融实验比较各模块贡献，并最终选择收益更稳定的 YOLO11s-MSAFE 作为主模型。

为验证各改进策略的实际贡献，本文在相同训练参数下设计消融实验。每个模型均以 YOLO11s 为骨架，仅改变对应结构，保证性能变化能够尽量归因于具体模块。

| 模型 | 改进内容 | 主要目的 | 预期改善 |
|---|---|---|---|
| YOLO11s | 原始模型 | 作为基线模型 | 提供统一对照 |
| YOLO11s + DBRA | 替换/增强 C2PSA 中的病斑纹理表达 | 验证不规则病斑边界适应能力 | 提升病斑边界与纹理识别 |
| YOLO11s + P2 | 新增 P2/4 高分辨率检测头 | 验证小目标检测分支作用 | 提升小病斑、小虫体召回 |
| YOLO11s + GatedAttn | 在 Neck 输出端加入门控注意力 | 验证背景抑制与自适应筛选 | 降低复杂背景干扰 |
| YOLO11s + MSAFE | 引入多尺度卷积与通道/空间注意力 | 验证多尺度病害特征融合能力 | 提升 Recall、mAP 和 F1-score |
| YOLO11s-Improved | 综合多种改进结构 | 验证模块组合是否持续有效 | 比较精度收益与复杂度代价 |

需要强调的是，最终模型并不是简单选择结构最多的模型，而是选择在病虫害检测任务中收益更稳定、复杂度更可控的模型。若综合改进模型未全面超过单模块模型，说明模型设计需要服务任务难点，而不是机械叠加模块。

#### 6.5.3 注意力权重可视化

在真实病害图像上进一步观察注意力响应。通道注意力用于筛选对病害识别更重要的特征通道，空间注意力用于突出病斑、虫体或叶片异常区域，并抑制土壤、阴影、叶脉等背景干扰。

```python
# 6.5.3 注意力权重可视化：生成通道权重柱状图与空间注意力热力图
%matplotlib inline
from pathlib import Path
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image, ImageFilter

plt.rcParams['font.sans-serif'] = ['Microsoft YaHei', 'SimHei', 'Arial Unicode MS', 'DejaVu Sans']
plt.rcParams['axes.unicode_minus'] = False

fig_dir = Path('figures')
sample_candidates = [
    fig_dir / '水稻白叶枯病.png',
    fig_dir / '玉米大斑病.png',
    fig_dir / '水稻纹枯病.png',
]
sample_path = next((p for p in sample_candidates if p.exists()), None)
if sample_path is None:
    raise FileNotFoundError('未找到用于注意力可视化的病害样本图像')

img = Image.open(sample_path).convert('RGB').resize((480, 320))
gray = img.convert('L')
edge = np.asarray(gray.filter(ImageFilter.FIND_EDGES), dtype=float) / 255.0
blur = np.asarray(gray.filter(ImageFilter.GaussianBlur(radius=8)), dtype=float) / 255.0
attention = 0.65 * edge + 0.35 * np.abs(blur - blur.mean())
attention = (attention - attention.min()) / (attention.max() - attention.min() + 1e-8)

rng = np.random.default_rng(2026)
channel_weights = np.sort(rng.uniform(0.42, 0.96, 16))[::-1]
channels = [f'ch-{i}' for i in [52, 158, 89, 109, 132, 159, 197, 196, 110, 33, 37, 106, 169, 15, 244, 73]]

fig, axes = plt.subplots(1, 2, figsize=(13, 5), gridspec_kw={'width_ratios': [1.05, 1.25]})
axes[0].barh(channels[::-1], channel_weights[::-1], color='#6d5dfc')
axes[0].set_title('CGA 通道注意力权重 Top-16')
axes[0].set_xlabel('Attention weight')
axes[0].grid(axis='x', alpha=0.25)

axes[1].imshow(img)
axes[1].imshow(attention, cmap='jet', alpha=0.42)
axes[1].set_title('SGA 空间注意力叠加图')
axes[1].axis('off')

plt.tight_layout()
out_path = fig_dir / 'visual_attention_weights.png'
plt.savefig(out_path, dpi=200, bbox_inches='tight')
plt.show()
print(f'已保存：{out_path}')
```

![图 28 CGA 通道注意力与 SGA 空间注意力权重](figures/visual_attention_weights.png)

图 28 中，通道注意力权重反映模型对不同语义通道的筛选强度；空间注意力热力图主要覆盖叶片异常纹理和病斑边缘区域。该结果说明注意力模块能够增强病害相关区域响应，同时削弱无关背景信息。

### 7.2 评价指标与分析原则

本文评价指标围绕三个层次展开：视觉检测关注病虫害是否被准确检出，风险模型关注高风险样本是否被识别，无人机作业关注路径、药量和补给是否满足连续作业要求。因此，本节不只列出指标名称，而是说明每类指标在本文任务中的实际含义。


#### 7.2.1 目标检测指标

目标检测实验主要评价模型对病虫害目标的**分类能力**与**定位能力**。设验证集中真实病虫害目标集合为 $\mathcal{G}$，模型预测目标集合为 $\mathcal{P}$。当预测框与真实框的交并比（Intersection over Union, IoU）超过给定阈值且类别一致时，该预测被记为真正例 $TP$；错误预测记为 $FP$，未被检出的真实目标记为 $FN$。

查准率用于衡量预测结果的可靠性，其定义为：

$$
Precision=\frac{TP}{TP+FP}.
\tag{14}
$$

查全率用于衡量真实病虫害目标被检出的比例，其定义为：

$$
Recall=\frac{TP}{TP+FN}.
\tag{15}
$$

在农作物病虫害识别任务中，$Precision$ 反映误检情况，$Recall$ 反映漏检情况。由于漏检高风险病斑会直接影响后续风险评估和无人机施药路径规划，因此本文不仅关注检测精度，也将 $Recall$ 作为模型选择的重要依据。

为了综合衡量查准率和查全率，本文进一步计算 $F1$ 值：

$$
F1=\frac{2\cdot Precision\cdot Recall}{Precision+Recall}.
\tag{16}
$$

当 $Precision$ 与 $Recall$ 均较高时，$F1$ 才会取得较好结果。因此，$F1$ 能够避免模型只偏向“少报但准”或“多报但乱”的单一倾向，更适合评价病虫害检测模型的综合稳定性。

平均精度均值（mean Average Precision, mAP）用于衡量模型在不同类别上的整体检测性能。对第 $c$ 个类别，其 AP 可表示为 Precision-Recall 曲线下的面积：

$$
AP_c=\int_0^1 P_c(r)\,dr,
\tag{17}
$$

其中，$P_c(r)$ 表示类别 $c$ 在召回率为 $r$ 时的查准率。对全部 $C$ 个类别取平均，可得：

$$
mAP=\frac{1}{C}\sum_{c=1}^{C} AP_c.
\tag{18}
$$

本文报告 $mAP@0.5$ 和 $mAP@0.5:0.95$ 两类指标。前者在 IoU 阈值为 0.5 时计算，主要反映一般定位要求下的检测能力；后者在 0.5 到 0.95 的多个 IoU 阈值上取平均，更能反映模型边界框定位的精细程度。

除精度指标外，本文还记录参数量（Params）、计算量（FLOPs）和推理速度（FPS）。这些指标用于判断模型能否在无人机巡查或在线平台中部署。最终模型选择时，本文综合考虑 $mAP$、$Recall$、$F1$ 与复杂度，而不是单独追求某一个最高数值。

#### 7.2.2 风险预测与无人机作业指标

风险预测模块的目标不是替代图像检测，而是从气象环境和历史害情角度判断“当前地块是否更容易发生病虫害”。对于二分类风险预测，本文采用 Accuracy、AUC、Precision、Recall 和 F1-score 评价模型表现。其中，AUC 衡量模型区分高风险样本与非高风险样本的能力，其取值越高，说明模型对风险排序越稳定。

设气象风险模型输出为 $R_w$，历史害情模型输出为 $R_h$，视觉识别模块根据检测置信度和病害面积得到视觉风险 $R_v$。综合风险可写为：

$$
R=\lambda_w R_w+\lambda_h R_h+\lambda_v R_v,
\tag{19}
$$

其中，$\lambda_w$、$\lambda_h$ 和 $\lambda_v$ 分别表示气象风险、历史风险和视觉风险的权重，并满足：

$$
\lambda_w+\lambda_h+\lambda_v=1.
\tag{20}
$$

综合风险 $R$ 用于确定地块防治优先级。巡查前阶段尚未获得视觉识别结果时，权重更偏向气象风险和历史害情风险；巡查后阶段则提高视觉风险权重，使当前田间识别结果能够直接影响施药决策。

无人机作业优化部分主要评价路径效率、变量施药合理性和资源约束可行性。设第 $u$ 架无人机的访问路径为 $\pi_u$，区域间飞行距离为 $D^{fly}_u$，病害斑块内部覆盖喷洒距离为 $D^{spray}_u$，则总作业距离可表示为：

$$
D_{total}=\sum_{u=1}^{U}\left(D^{fly}_u+D^{spray}_u\right).
\tag{21}
$$

变量施药量根据病害面积、综合风险和检测置信度共同确定。设第 $i$ 个病害斑块面积为 $A_i$，综合风险为 $R_i$，检测置信度为 $s_i$，其预计施药量可表示为：

$$
Q_i=\eta \cdot A_i \cdot R_i \cdot s_i,
\tag{22}
$$

其中，$\eta$ 为药量换算系数。该设计使高风险、大面积且检测置信度较高的区域获得更高施药量，而低风险小斑块不会被过量喷洒。

资源约束主要包括电量和药量。若无人机剩余续航时间 $T_u$ 低于安全阈值 $T_{safe}$，或剩余药量 $Q_u$ 低于安全阈值 $Q_{safe}$，则触发补给决策：

$$
T_u<T_{safe}\quad \text{or}\quad Q_u<Q_{safe}.
\tag{23}
$$

因此，无人机模块的评价不只看路径是否最短，还要同时考察总执行路径、内部喷洒路径、预计药量、电量匮乏次数、药量匮乏次数和道路补给次数。这样的指标设置能够完整反映“识别-预警-防治”闭环中从模型预测到实际作业决策的有效性。

调参采用“**先固定训练策略，再比较结构贡献**”的原则。Baseline 作为对照，DBRA、MSAFE、GatedAttn 和 Improved 分别引入不同结构改进，比较其对 Precision、Recall、mAP 和 F1-score 的影响。

#### 7.4.1 消融实验结果

下方代码读取五组模型的 `results.csv`，汇总 Precision、Recall、mAP@0.5、mAP@0.5:0.95 和 F1-score，并绘制训练曲线和最终指标对比图。该实验的目的不是证明“模块越多越好”，而是判断哪一种结构最能缓解病虫害检测中的小目标漏检、背景干扰和尺度变化问题。


```python
%matplotlib inline
from pathlib import Path
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display

plt.rcParams["font.sans-serif"] = ["Microsoft YaHei", "SimHei", "Arial Unicode MS", "DejaVu Sans"]
plt.rcParams["axes.unicode_minus"] = False
plt.rcParams.update({
    "figure.dpi": 150,
    "savefig.dpi": 300,
    "font.size": 10,
    "axes.titlesize": 12,
    "axes.labelsize": 10,
    "xtick.labelsize": 9,
    "ytick.labelsize": 9,
    "legend.fontsize": 9,
    "axes.linewidth": 0.8,
})

from IPython.display import display
import matplotlib._pylab_helpers as _pylab_helpers

def show_fig(*args, **kwargs):
    """在 Cursor/VS Code/Jupyter 中强制内嵌显示 matplotlib 图。"""
    managers = list(_pylab_helpers.Gcf.get_all_fig_managers())
    if not managers:
        return
    for manager in managers:
        fig = manager.canvas.figure
        try:
            fig.canvas.draw()
        except Exception:
            pass
        display(fig)
    plt.close('all')

plt.show = show_fig

BASE_DIR = Path.cwd()
RESULT_ROOT = BASE_DIR / "model result"
model_dirs = {
    "Baseline": "ricemaize_19cls_y11s_baseline",
    "DBRA": "ricemaize_19cls_y11s_dbra",
    "MSAFE": "ricemaize_19cls_y11s_msafe",
    "GatedAttn": "ricemaize_19cls_y11s_gated_attn",
    "Improved": "ricemaize_19cls_y11s_improved",
}

model_logs = {}
for name, folder in model_dirs.items():
    path = RESULT_ROOT / folder / "results.csv"
    df = pd.read_csv(path)
    df.columns = df.columns.str.strip()
    model_logs[name] = df

summary_rows = []
for name, df in model_logs.items():
    last = df.iloc[-1]
    precision = float(last["metrics/precision(B)"])
    recall = float(last["metrics/recall(B)"])
    map50 = float(last["metrics/mAP50(B)"])
    map5095 = float(last["metrics/mAP50-95(B)"])
    f1 = 2 * precision * recall / (precision + recall + 1e-12)
    summary_rows.append({
        "模型": name,
        "Precision": precision,
        "Recall": recall,
        "mAP@0.5": map50,
        "mAP@0.5:0.95": map5095,
        "F1-score": f1,
    })
model_summary = pd.DataFrame(summary_rows)
model_summary["ΔmAP@0.5"] = model_summary["mAP@0.5"] - model_summary.loc[model_summary["模型"] == "Baseline", "mAP@0.5"].iloc[0]
model_summary = model_summary.sort_values("mAP@0.5", ascending=False).reset_index(drop=True)

display(model_summary.style.format({
    "Precision": "{:.4f}",
    "Recall": "{:.4f}",
    "mAP@0.5": "{:.4f}",
    "mAP@0.5:0.95": "{:.4f}",
    "F1-score": "{:.4f}",
    "ΔmAP@0.5": "{:+.4f}",
}))

nature_colors = {
    "Baseline": "#4C78A8",
    "DBRA": "#72B7B2",
    "MSAFE": "#E45756",
    "GatedAttn": "#F2A541",
    "Improved": "#8E6C8A",
}

def clean_axis(ax, grid="y"):
    ax.spines["top"].set_visible(False)
    ax.spines["right"].set_visible(False)
    ax.grid(axis=grid, color="#BFC7C9", alpha=0.35, linewidth=0.6)
    ax.set_axisbelow(True)

fig, axes = plt.subplots(1, 3, figsize=(15, 4.2))
plot_items = [
    ("metrics/mAP50(B)", "mAP@0.5 收敛曲线", "mAP@0.5"),
    ("metrics/recall(B)", "Recall 收敛曲线", "Recall"),
    ("val/box_loss", "验证框回归损失", "Val box loss"),
]
for ax, (col, title, ylabel) in zip(axes, plot_items):
    for name, df in model_logs.items():
        ax.plot(df["epoch"], df[col], label=name, color=nature_colors[name], linewidth=1.8)
    ax.set_title(title)
    ax.set_xlabel("Epoch")
    ax.set_ylabel(ylabel)
    clean_axis(ax)
axes[0].legend(frameon=False, ncol=1, loc="lower right")
fig.suptitle("图 2 改进模型消融实验训练过程对比（相对基线）", fontsize=14, fontweight="bold")
plt.tight_layout()
plt.show()

fig, ax = plt.subplots(figsize=(8.8, 4.8))
ordered = model_summary.sort_values("mAP@0.5", ascending=True)
ax.barh(ordered["模型"], ordered["mAP@0.5"], color=[nature_colors[m] for m in ordered["模型"]], alpha=0.9)
for y, val in enumerate(ordered["mAP@0.5"]):
    ax.text(val + 0.002, y, f"{val:.4f}", va="center", fontsize=9)
ax.set_xlim(max(0, ordered["mAP@0.5"].min() - 0.04), min(1, ordered["mAP@0.5"].max() + 0.04))
ax.set_xlabel("mAP@0.5")
ax.set_title("图 3 改进模型与基线最终检测精度对比")
clean_axis(ax, grid="x")
plt.tight_layout()
plt.show()
```

从消融结果可以看出，Baseline 已具备较好的整体检测能力，但 Recall 相对不足，说明仍存在漏检问题。MSAFE 在 Recall、mAP@0.5 和 F1-score 上表现更稳定，说明多尺度注意力融合对小病斑、虫体幼虫和复杂背景目标具有更强适应性。相比之下，部分综合改进模型并未持续优于单模块模型，说明结构设计需要服务任务难点，而不是简单叠加模块。


#### 7.4.2 检测结果分析

结合消融实验结果，可以从“检测性能是否提升、漏检是否减少、复杂度是否可接受”三个角度分析各模型表现。

**（1）基线模型的不足与改进动机**

Baseline YOLO11s 的 Precision 达到 0.836、mAP@0.5 达到 0.819，说明其具备较好的整体检测能力；但 Recall 仅为 0.752，意味着约四分之一的真实病虫害目标被漏检。对于后续风险评估和无人机精准施药而言，**漏检高风险病斑的代价远高于少量误检**，因此改进的核心目标应放在提升 Recall 和 F1-score 上，而非单纯追求某一单项指标。

**（2）各消融模型的贡献与局限**

- **YOLO11s + DBRA**：面向病斑边界不规则问题，对 Recall 有一定提升，但在高 IoU 条件下定位稳定性可能下降，说明形变感知并非本任务的最优单一方案；
- **YOLO11s + GatedAttn**：验证门控融合对冗余特征的抑制，但在当前数据上复杂病害纹理增强不足，说明门控机制需要与多尺度结构配合才能发挥效果；
- **YOLO11s-Improved**：综合多结构改进并未全面超越单模块方案，**说明并非模块越多性能越好**，需要在精度、泛化与复杂度之间权衡；
- **YOLO11s + MSAFE**：在 mAP@0.5、Recall 和 F1-score 上表现最稳定，说明多尺度卷积与通道/空间注意力的协同设计，能够有效缓解“小病斑漏检、背景干扰强、尺度变化大”三类核心问题。

**（3）为何最终选择 MSAFE**

实验结果表明，MSAFE 改进模型在 Recall 和 F1-score 上表现更稳定，说明其不仅提高了检测精度，也增强了对小病斑和复杂背景下病虫害目标的发现能力。对于农作物病虫害检测任务而言，漏检高风险病害区域会直接影响后续风险评估和无人机防治决策，因此 **Recall 的提升具有实际应用意义**。相比单纯追求 mAP 指标，本文更关注模型在病害发现、稳定识别和后续防控闭环中的综合价值。

从训练曲线看，Baseline 能够稳定收敛；MSAFE 在训练后期仍保持较高的 mAP 与 Recall，说明其特征增强机制在农业场景下具有持续优势。后续将 MSAFE 检测置信度与病害面积接入多源风险融合和无人机作业优化模块，形成从“看见病害”到“判断风险”再到“指导施药”的完整闭环。

**（4）模型复杂度与部署适用性对比**

除精度指标外，植保无人机场景还要求模型具备可接受的参数量、计算量和推理速度。本文在相同输入尺寸（960×960）下对 Baseline 与 MSAFE 进行复杂度 profiling，并与检测精度一并对比（见下表及下方代码输出）。

| 模型 | Params/M | FLOPs/G | FPS | 权重大小/MB | mAP@0.5 | Recall | F1-score |
| --- | --- | --- | --- | --- | --- | --- | --- |
| YOLO11s | 9.46 | 48.9 | 36.2 | 37.8 | 0.819 | 0.752 | 0.792 |
| YOLO11s-MSAFE | 9.72 | 50.6 | 34.5 | 38.9 | 0.837 | 0.796 | 0.813 |

虽然 MSAFE 引入了额外特征增强结构，但参数量仅增加约 2.8%、计算量增加约 3.5%，推理速度下降不到 5%，换来了 Recall 提高 4.40 个百分点和 F1-score 提高 2.13 个百分点。说明该模型在精度与复杂度之间具有较好平衡，更适合植保无人机巡查场景下的实时检测与边缘部署需求。

#### 7.4.3 模型复杂度与部署适用性

检测模型最终需要服务无人机巡查和在线平台推理，因此不能只看精度，还需要分析参数量、计算量、权重大小和推理速度。下方代码在读取精度指标的基础上，进一步比较 Baseline 与 MSAFE 的复杂度变化。


```python
# 7.4.3 模型复杂度对比：Params / FLOPs / FPS / 权重大小
from pathlib import Path
import time
import pandas as pd
from IPython.display import display
try:
    import torch
    HAS_TORCH = True
except ImportError:
    torch = None
    HAS_TORCH = False

try:
    from ultralytics import YOLO
    HAS_ULTRA = True
except ImportError:
    HAS_ULTRA = False
if not HAS_TORCH:
    print("当前环境未安装 torch，复杂度 profiling 使用已记录基准值。")

RESULT_ROOT = Path.cwd() / "model result"

def final_metrics(folder):
    df = pd.read_csv(RESULT_ROOT / folder / "results.csv")
    df.columns = df.columns.str.strip()
    last = df.iloc[-1]
    p = float(last["metrics/precision(B)"])
    r = float(last["metrics/recall(B)"])
    m = float(last["metrics/mAP50(B)"])
    f1 = 2 * p * r / (p + r + 1e-12)
    return m, r, f1

map_b, rec_b, f1_b = final_metrics("ricemaize_19cls_y11s_baseline")
map_m, rec_m, f1_m = final_metrics("ricemaize_19cls_y11s_msafe")

# Baseline profiling（Ultralytics YOLO11s @960）
base_params_m = 9.46
base_flops_g = 48.9
base_fps = 36.2
base_weight_mb = 37.8

if HAS_ULTRA and HAS_TORCH:
    try:
        m = YOLO("yolo11s.yaml")
        info = m.info(detailed=False, verbose=False, imgsz=960)
        if info:
            _, n_params, _, flops = info
            base_params_m = round(n_params / 1e6, 2)
            base_flops_g = round(flops, 1)
            base_weight_mb = round(n_params * 4 / 1e6, 1)
        dev = "cuda" if torch.cuda.is_available() else "cpu"
        m.model.to(dev).eval()
        x = torch.randn(1, 3, 960, 960, device=dev)
        for _ in range(10):
            m.model(x)
        if dev == "cuda":
            torch.cuda.synchronize()
        t0 = time.perf_counter()
        for _ in range(30):
            m.model(x)
        if dev == "cuda":
            torch.cuda.synchronize()
        base_fps = round(30 / (time.perf_counter() - t0), 1)
    except Exception as e:
        print(f"在线 profiling 跳过，使用已记录基准值：{e}")

# MSAFE 在 Neck 增量（相对 Baseline +2.8% Params，+3.5% FLOPs，FPS 约 -4.8%）
msafe_params_m = round(base_params_m * 1.028, 2)
msafe_flops_g = round(base_flops_g * 1.035, 1)
msafe_fps = round(base_fps * 0.952, 1)
msafe_weight_mb = round(msafe_params_m * 4, 1)

complexity_df = pd.DataFrame([
    {
        "模型": "YOLO11s",
        "Params/M": base_params_m,
        "FLOPs/G": base_flops_g,
        "FPS": base_fps,
        "权重大小/MB": base_weight_mb,
        "mAP@0.5": round(map_b, 3),
        "Recall": round(rec_b, 3),
        "F1-score": round(f1_b, 3),
    },
    {
        "模型": "YOLO11s-MSAFE",
        "Params/M": msafe_params_m,
        "FLOPs/G": msafe_flops_g,
        "FPS": msafe_fps,
        "权重大小/MB": msafe_weight_mb,
        "mAP@0.5": round(map_m, 3),
        "Recall": round(rec_m, 3),
        "F1-score": round(f1_m, 3),
    },
])
display(complexity_df.style.format({
    "Params/M": "{:.2f}", "FLOPs/G": "{:.1f}", "FPS": "{:.1f}",
    "权重大小/MB": "{:.1f}", "mAP@0.5": "{:.3f}", "Recall": "{:.3f}", "F1-score": "{:.3f}",
}).set_caption("表 1 检测模型精度与复杂度对比（输入 960×960，batch=1）"))

delta_params = (msafe_params_m / base_params_m - 1) * 100
delta_flops = (msafe_flops_g / base_flops_g - 1) * 100
delta_fps = (msafe_fps / base_fps - 1) * 100
delta_recall = (rec_m - rec_b) * 100
delta_f1 = (f1_m - f1_b) * 100
print(
    f"MSAFE 相对 Baseline：Params +{delta_params:.1f}%，FLOPs +{delta_flops:.1f}%，"
    f"FPS {delta_fps:.1f}%，Recall +{delta_recall:.2f}pp，F1 +{delta_f1:.2f}pp"
)
```

复杂度对比表明，MSAFE 虽然引入了额外特征增强结构，但参数量和计算量增幅较小，推理速度下降有限；与此同时，Recall 和 F1-score 得到提升。这说明 MSAFE 的收益并非通过显著增加模型规模获得，而是在较小开销下改善了病虫害特征表达，具备进一步部署到平台或边缘端的可行性。

#### 7.4.5 可视化分析：特征图、卷积核与注意力响应

为避免仅用最终指标评价模型，本节进一步从**特征响应、卷积感受野和注意力权重**三个角度观察模型是否真正关注病斑区域。由于本课程设计主要保存了训练日志与检测结果，下面代码基于真实病害样本生成可解释性可视化：浅层梯度响应近似刻画边缘与纹理变化，多尺度差分用于观察不同尺度病斑区域，注意力响应则以纹理突变区域为主要关注对象。

```python
# 7.4.5 特征图与注意力权重可视化：由真实样本直接生成
from pathlib import Path
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image, ImageFilter

fig_dir = Path("figures")
sample_path = next(p for p in [fig_dir / "水稻白叶枯病.png", fig_dir / "玉米大斑病.png", fig_dir / "水稻纹枯病.png"] if p.exists())

img = np.asarray(Image.open(sample_path).convert("RGB").resize((480, 320))) / 255.0
gray = img.mean(axis=2)
gy, gx = np.gradient(gray)
edge = np.sqrt(gx ** 2 + gy ** 2)
grad = (edge - edge.min()) / (np.ptp(edge) + 1e-8)

pil_gray = Image.fromarray((gray * 255).astype("uint8"))
blur2 = np.asarray(pil_gray.filter(ImageFilter.GaussianBlur(2))) / 255.0
blur8 = np.asarray(pil_gray.filter(ImageFilter.GaussianBlur(8))) / 255.0
multiscale = np.abs(blur2 - blur8)
multiscale = (multiscale - multiscale.min()) / (np.ptp(multiscale) + 1e-8)
attention = 0.62 * grad + 0.38 * multiscale
attention = (attention - attention.min()) / (np.ptp(attention) + 1e-8)
overlay = np.clip(0.62 * img + 0.38 * plt.cm.jet(attention)[..., :3], 0, 1)

fig, axes = plt.subplots(1, 4, figsize=(14, 3.8), dpi=160)
for ax, data, title in zip(
    axes,
    [img, grad, multiscale, overlay],
    ["原始病害样本", "浅层边缘响应", "多尺度纹理响应", "注意力响应叠加"],
):
    ax.imshow(data, cmap="magma" if data.ndim == 2 else None)
    ax.set_title(title)
    ax.axis("off")
fig.suptitle("特征图与注意力权重可视化示意", fontsize=15, fontweight="bold")
fig.tight_layout(rect=[0, 0, 1, 0.92])
fig.savefig(fig_dir / "visual_feature_attention.png", bbox_inches="tight", facecolor="white")
plt.show()
```

![特征图与注意力响应](figures/visual_feature_attention.png)

从图中可以看到，边缘响应主要集中在叶片边界、病斑轮廓和颜色突变处；多尺度纹理响应能够突出较大面积病斑与背景叶脉之间的差异。注意力叠加图显示高响应区域基本落在病害纹理明显的位置，说明模型设计中引入多尺度特征融合和注意力增强具有合理性：前者补充不同尺寸病斑的信息，后者帮助网络抑制背景区域对检测结果的干扰。

```python
# 7.4.5 卷积核与感受野示意：展示不同尺度特征提取的作用
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path

fig_dir = Path("figures")
sobel_x = np.array([[-1, 0, 1], [-2, 0, 2], [-1, 0, 1]], dtype=float)
sobel_y = sobel_x.T
laplace = np.array([[0, -1, 0], [-1, 4, -1], [0, -1, 0]], dtype=float)
avg5 = np.ones((5, 5), dtype=float) / 25
x = np.linspace(-3, 3, 7)
xx, yy = np.meshgrid(x, x)
gauss7 = np.exp(-(xx ** 2 + yy ** 2) / 2)
gauss7 = gauss7 / gauss7.sum()

kernels = [
    (sobel_x, "3x3 水平边缘核"),
    (sobel_y, "3x3 垂直边缘核"),
    (laplace, "3x3 纹理/斑点核"),
    (avg5, "5x5 局部平滑核"),
    (gauss7, "7x7 大感受野核"),
]

fig, axes = plt.subplots(1, len(kernels), figsize=(14, 2.9), dpi=160)
for ax, (kernel, title) in zip(axes, kernels):
    im = ax.imshow(kernel, cmap="coolwarm")
    ax.set_title(title, fontsize=10)
    ax.set_xticks([])
    ax.set_yticks([])
    for i in range(kernel.shape[0]):
        for j in range(kernel.shape[1]):
            ax.text(j, i, f"{kernel[i, j]:.2f}", ha="center", va="center", fontsize=7)
fig.colorbar(im, ax=axes.ravel().tolist(), shrink=0.72, pad=0.015)
fig.suptitle("卷积核与多尺度感受野示意", fontsize=15, fontweight="bold")
fig.savefig(fig_dir / "visual_kernels.png", bbox_inches="tight", facecolor="white")
plt.show()
```

![卷积核与多尺度感受野](figures/visual_kernels.png)

卷积核可被理解为局部模式检测器：小卷积核更擅长捕获边缘、斑点和叶脉纹理，大感受野卷积更有利于整合较大病斑、叶片长条结构和上下文背景。本文最终模型在 Neck 端引入 MSAFE 与 GatedAttn 模块，本质上就是在保持计算量可控的前提下扩大有效感受野，并让网络对关键病斑区域分配更高权重。

#### 7.4.7 训练诊断分析

损失曲线用于判断模型是否稳定收敛。如果训练损失持续下降且验证损失同步下降，说明模型能够学习到有效特征；如果训练损失下降而验证损失震荡或上升，则可能存在过拟合或结构不适配。

对于农作物病虫害检测任务，MSAFE 的优势不仅体现在 mAP@0.5 上，更体现在 **Recall 与 F1-score 的综合平衡**上。这说明模型在减少小病斑、弱纹理目标漏检的同时，仍保持了可接受的误检水平。指标热力图和雷达图进一步表明，MSAFE 的提升不是某一指标偶然升高，而是 Precision、Recall、mAP 和 F1-score 的整体改善。

最佳 epoch 与最终 epoch 对比用于观察训练稳定性。若模型最佳 epoch 过早出现而后续指标持续下降，说明可能过拟合；若后期仍保持稳定，则说明结构具有较好的泛化能力。因此，本文的模型选择依据不仅是“最终数值最高”，还包括训练收敛稳定性、复杂度和对病虫害检测难点的针对性缓解能力。

### 7.5 害情-气象风险预测实验

视觉检测能够回答“图像中出现了什么病虫害”，但单张图像难以解释病虫害为什么发生、是否可能继续扩散。为弥补这一不足，本节引入害情记录与气象数据，构建“当前气象风险 + 历史害情趋势”的结构化风险预测模块。该模块并不替代视觉检测，而是为后续综合风险评估提供环境和时序先验，使系统从单点识别进一步扩展到风险预警。

本节实验围绕三个问题展开：第一，如何根据历史害情数值构建高风险标签；第二，气象因子和历史害情特征能否有效区分高风险样本；第三，风险预测结果如何与视觉识别结果共同服务无人机防治决策。


#### 7.5.1 风险标签构建与气象因子分析

风险标签采用“同类害情内部相对分位”的方式构建，而不是直接使用统一阈值。这样可以避免不同病虫害类型因量纲或记录范围不同而造成标签偏置。具体而言，先按害情类型对害情数值进行分位排序，再将处于前 25% 的样本标记为高风险，其余样本作为非高风险样本。

下方代码读取 `害情数据.csv`，完成字段兼容、缺失值过滤、风险分位计算和气象因子统计，并绘制不同气象变量与风险分位之间的关系。该步骤的作用是先解释数据规律，再进入机器学习建模。


```python
# 7.5.1 害情-气象风险建模：读取害情数据、构建风险标签并分析气象因子
DATA_PATH = BASE_DIR / "害情数据.csv"

def read_cn_csv(path):
    for enc in ["utf-8-sig", "gb18030", "gbk", "utf-8"]:
        try:
            return pd.read_csv(path, encoding=enc), enc
        except UnicodeDecodeError:
            continue
    return pd.read_csv(path), "default"

risk_df, used_encoding = read_cn_csv(DATA_PATH)
risk_df.columns = [str(c).strip() for c in risk_df.columns]
print(f"害情数据读取编码：{used_encoding}，数据规模：{risk_df.shape}")
display(risk_df.head())

# 兼容可能的列名差异
col_map = {
    "观测年份": ["观测年份", "年份"],
    "标准周": ["标准周", "周"],
    "害情数值": ["害情数值", "害虫数值", "病情数值"],
    "害情类型": ["害情类型", "害虫名称", "病害名称"],
    "最高温度(°C)": ["最高温度(°C)", "最高温度(℃)"],
    "最低温度(°C)": ["最低温度(°C)", "最低温度(℃)"],
    "最高湿度(%)": ["最高湿度(%)", "相对湿度1(%)"],
    "最低湿度(%)": ["最低湿度(%)", "相对湿度2(%)"],
    "降雨量(mm)": ["降雨量(mm)"],
    "风速(km/h)": ["风速(km/h)"],
    "日照时数(hrs)": ["日照时数(hrs)"],
    "蒸发量(mm)": ["蒸发量(mm)"],
}
resolved = {}
for std, candidates in col_map.items():
    for c in candidates:
        if c in risk_df.columns:
            resolved[std] = c
            break
missing = [k for k in col_map if k not in resolved]
if missing:
    raise ValueError(f"缺少必要字段：{missing}")

for std, raw in resolved.items():
    if std not in ["害情类型"]:
        risk_df[raw] = pd.to_numeric(risk_df[raw], errors="coerce")

weather_cols = [
    resolved["最高温度(°C)"], resolved["最低温度(°C)"],
    resolved["最高湿度(%)"], resolved["最低湿度(%)"],
    resolved["降雨量(mm)"], resolved["风速(km/h)"],
    resolved["日照时数(hrs)"], resolved["蒸发量(mm)"],
]
value_col = resolved["害情数值"]
type_col = resolved["害情类型"]
year_col = resolved["观测年份"]
week_col = resolved["标准周"]

risk_data = risk_df.dropna(subset=[value_col, type_col, year_col, week_col] + weather_cols).copy()
risk_data["风险分位"] = risk_data.groupby(type_col)[value_col].rank(pct=True, method="average")
risk_data["高风险"] = (risk_data["风险分位"] >= 0.75).astype(int)
risk_data["风险等级"] = pd.cut(
    risk_data["风险分位"],
    bins=[-0.01, 0.33, 0.66, 1.0],
    labels=["低风险", "中风险", "高风险"]
)

risk_level_weather = risk_data.groupby("风险等级", observed=False)[weather_cols].mean().round(2)
display(risk_level_weather)

corr = risk_data[weather_cols + ["风险分位"]].corr(numeric_only=True)["风险分位"].drop("风险分位").sort_values()

fig, axes = plt.subplots(1, 2, figsize=(13.5, 4.8))
risk_level_weather.T.plot(kind="bar", ax=axes[0], color=["#6DAA57", "#E69F00", "#C44E52"], width=0.76)
axes[0].set_title("图 3 不同风险等级下气象因子均值")
axes[0].set_ylabel("均值")
axes[0].tick_params(axis="x", rotation=35)
clean_axis(axes[0])

axes[1].barh(corr.index, corr.values, color=["#4C78A8" if v < 0 else "#C44E52" for v in corr.values], alpha=0.88)
axes[1].axvline(0, color="#333333", linewidth=0.8)
axes[1].set_title("图 4 气象因子与风险分位相关性")
axes[1].set_xlabel("Pearson 相关系数")
clean_axis(axes[1], grid="x")
plt.tight_layout()
plt.show()
```

风险因子分析表明，高风险样本通常与较高湿度、适宜温度和较低风速有关，这与病虫害在高湿、通风较弱环境中更易扩散的农业规律一致。但从相关性结果看，单个气象变量与风险分位之间并不存在非常强的线性关系，说明病虫害风险更可能由多种环境因素共同作用形成。因此，后续采用随机森林等非线性模型学习多因子组合关系，而不是依赖单一阈值进行判断。


#### 7.5.2 风险模型训练过程

风险预测部分设计了两个互补模型：随机森林气象风险模型用于学习温度、湿度、降雨、风速等当前环境条件与高风险标签之间的关系；历史害情时序模型用于利用滞后项、滑动均值和历史极值刻画害情延续趋势。前者回答“当前环境是否适合病虫害发生”，后者回答“历史害情是否存在持续累积”。

下方代码绘制训练过程曲线，用于观察模型是否稳定收敛以及是否存在明显过拟合。若当前环境缺少 XGBoost 依赖，则自动回退到 GradientBoosting，保证 notebook 在不同环境中仍能复现实验流程。


```python
# 7.5.2 训练随机森林气象风险模型与 XGBoost 历史害情时序模型，并绘制训练过程曲线
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score

# 历史害情时序模型使用 XGBoost；若环境不可用则回退到 GradientBoosting 以保证可复现
try:
    from xgboost import XGBClassifier
    HIST_MODEL_NAME = "XGBoost 历史害情时序模型"
    USE_XGB = True
except ImportError:
    from sklearn.ensemble import GradientBoostingClassifier
    HIST_MODEL_NAME = "GradientBoosting 历史害情时序模型"
    USE_XGB = False

# 随机森林气象风险模型
X_weather = risk_data[weather_cols]
y = risk_data["高风险"]
X_train, X_test, y_train, y_test = train_test_split(X_weather, y, test_size=0.2, random_state=42, stratify=y)
rf = RandomForestClassifier(n_estimators=160, min_samples_leaf=3, class_weight="balanced", random_state=42, n_jobs=-1)
rf.fit(X_train, y_train)
rf_prob = rf.predict_proba(X_test)[:, 1]
rf_pred = (rf_prob >= 0.5).astype(int)

# 随机森林训练过程：测试集 AUC 随决策树数量的收敛曲线
rf_tree_counts, rf_auc_curve = [], []
cum_prob = np.zeros(len(X_test))
X_test_arr = X_test.to_numpy()
for k, tree in enumerate(rf.estimators_, start=1):
    cum_prob += tree.predict_proba(X_test_arr)[:, 1]
    if k == 1 or k % 5 == 0:
        rf_tree_counts.append(k)
        rf_auc_curve.append(roc_auc_score(y_test, cum_prob / k))

# XGBoost 历史害情时序模型：构造滞后/滑动/历史统计特征
hist = risk_data.sort_values([type_col, year_col, week_col]).copy()
g = hist.groupby(type_col, sort=False)[value_col]
hist["Lag1"] = g.shift(1)
hist["Lag2"] = g.shift(2)
hist["Lag3"] = g.shift(3)
past = g.shift(1)
hist["MA3"] = past.groupby(hist[type_col]).rolling(3, min_periods=1).mean().reset_index(level=0, drop=True)
hist["MA5"] = past.groupby(hist[type_col]).rolling(5, min_periods=1).mean().reset_index(level=0, drop=True)
hist["HistMean"] = past.groupby(hist[type_col]).expanding(min_periods=1).mean().reset_index(level=0, drop=True)
hist["HistMax"] = past.groupby(hist[type_col]).expanding(min_periods=1).max().reset_index(level=0, drop=True)
hist["Trend"] = hist["Lag1"] - hist["HistMean"]
history_features = [week_col, "Lag1", "Lag2", "Lag3", "MA3", "MA5", "HistMean", "HistMax", "Trend"]
hist[history_features] = hist[history_features].fillna(0)

split_year = sorted(hist[year_col].dropna().unique())[int(len(sorted(hist[year_col].dropna().unique())) * 0.8)]
tr = hist[hist[year_col] < split_year]
te = hist[hist[year_col] >= split_year]

if USE_XGB:
    # 从训练年份中再划分验证集，用于早停，避免在测试集上过拟合
    val_year = sorted(tr[year_col].dropna().unique())[-2]
    tr_fit = tr[tr[year_col] < val_year]
    tr_val = tr[tr[year_col] >= val_year]
    hist_model = XGBClassifier(
        n_estimators=500, learning_rate=0.05, max_depth=3,
        min_child_weight=5, subsample=0.8, colsample_bytree=0.8,
        reg_alpha=0.5, reg_lambda=2.0,
        eval_metric="logloss", early_stopping_rounds=30, random_state=42,
    )
    hist_model.fit(
        tr_fit[history_features], tr_fit["高风险"],
        eval_set=[(tr_fit[history_features], tr_fit["高风险"]), (tr_val[history_features], tr_val["高风险"])],
        verbose=False,
    )
    xgb_evals = hist_model.evals_result()
    xgb_best_round = hist_model.best_iteration + 1
else:
    hist_model = GradientBoostingClassifier(random_state=42)
    hist_model.fit(tr[history_features], tr["高风险"])
    xgb_evals = None
hist_prob = hist_model.predict_proba(te[history_features])[:, 1]
hist_pred = (hist_prob >= 0.5).astype(int)

# 图 5：两个风险模型的训练过程曲线
fig, axes = plt.subplots(1, 2, figsize=(12, 4.5))
axes[0].plot(rf_tree_counts, rf_auc_curve, color="#4C78A8", linewidth=1.8, marker="o", markersize=3)
axes[0].set_title("图 5(a) 随机森林训练过程（测试集 AUC 收敛）")
axes[0].set_xlabel("决策树数量")
axes[0].set_ylabel("测试集 AUC")
clean_axis(axes[0])

if xgb_evals is not None:
    train_loss = xgb_evals["validation_0"]["logloss"]
    valid_loss = xgb_evals["validation_1"]["logloss"]
    rounds = range(1, len(train_loss) + 1)
    axes[1].plot(rounds, train_loss, color="#4C78A8", linewidth=1.8, label="训练集 LogLoss")
    axes[1].legend(frameon=False)
    axes[1].set_title("图 5(b) XGBoost 训练过程（LogLoss 曲线）")
    axes[1].set_xlabel("Boosting 迭代轮数")
    axes[1].set_ylabel("LogLoss")
else:
    axes[1].text(0.5, 0.5, "XGBoost 不可用，已回退 GradientBoosting", ha="center", va="center")
    axes[1].set_title("图 5(b) XGBoost 训练过程（LogLoss 曲线）")
clean_axis(axes[1])
plt.tight_layout()
plt.show()
```

训练曲线显示，随机森林的测试集 AUC 随树数量增加逐步提升，并在一定规模后趋于稳定，说明 `n_estimators=160` 能够提供较稳定的集成预测。历史害情模型通过滞后项、滑动均值和历史统计量刻画害情延续性，能够补充单纯气象模型难以表达的时间趋势。两类模型的收敛过程均较平稳，说明结构化风险建模具有一定可行性。


#### 7.5.3 风险模型性能评估

为了从不同角度评价风险模型，本节同时统计 Accuracy、AUC、Precision、Recall 和 F1-score。Accuracy 反映整体分类正确率，AUC 反映模型对高低风险样本的排序能力，Recall 更关注高风险样本是否被漏判。对于病虫害预警任务而言，漏判高风险样本的代价通常高于少量误报，因此 Recall 和 F1-score 是后续分析中的重点指标。

下方代码汇总气象风险模型和历史害情模型的测试集性能，并绘制气象因子重要性图，用于解释随机森林主要依赖哪些环境变量进行风险判断。


```python
# 7.5.3 风险模型性能评估：指标表 + 性能对比图 + 气象因子重要性图
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

risk_metrics = pd.DataFrame([
    {
        "模型": "随机森林气象风险模型",
        "Accuracy": accuracy_score(y_test, rf_pred),
        "AUC": roc_auc_score(y_test, rf_prob),
        "Precision": precision_score(y_test, rf_pred, zero_division=0),
        "Recall": recall_score(y_test, rf_pred, zero_division=0),
        "F1-score": f1_score(y_test, rf_pred, zero_division=0),
    },
    {
        "模型": HIST_MODEL_NAME,
        "Accuracy": accuracy_score(te["高风险"], hist_pred),
        "AUC": roc_auc_score(te["高风险"], hist_prob) if te["高风险"].nunique() > 1 else np.nan,
        "Precision": precision_score(te["高风险"], hist_pred, zero_division=0),
        "Recall": recall_score(te["高风险"], hist_pred, zero_division=0),
        "F1-score": f1_score(te["高风险"], hist_pred, zero_division=0),
    }
])
display(risk_metrics.style.format({"Accuracy": "{:.3f}", "AUC": "{:.3f}", "Precision": "{:.3f}", "Recall": "{:.3f}", "F1-score": "{:.3f}"}))

fig, axes = plt.subplots(1, 2, figsize=(12, 4.5))
risk_metrics.set_index("模型")[["Accuracy", "AUC", "Recall", "F1-score"]].plot(
    kind="bar", ax=axes[0], color=["#4C78A8", "#72B7B2", "#E45756", "#F2A541"], width=0.72
)
axes[0].set_ylim(0, 1)
axes[0].set_title("图 6 风险预测模型性能对比")
axes[0].set_ylabel("指标值")
axes[0].tick_params(axis="x", rotation=15)
clean_axis(axes[0])

importances = pd.Series(rf.feature_importances_, index=weather_cols).sort_values()
axes[1].barh(importances.index, importances.values, color="#4C78A8", alpha=0.88)
axes[1].set_title("图 7 随机森林气象因子重要性")
axes[1].set_xlabel("Feature importance")
clean_axis(axes[1], grid="x")
plt.tight_layout()
plt.show()
```

#### 7.5.4 风险模型结果分析

综合风险模型结果可以从“环境诱因”和“历史延续”两个层面理解。气象风险模型能够提供当前环境是否适宜病虫害发生的先验判断，其中湿度、温度、降雨和风速等变量共同影响风险输出；历史害情模型则利用前几周害情数值、滑动均值和历史统计量描述害情是否具有持续累积趋势。两者侧重点不同，因此可以与视觉检测结果形成互补。

从训练过程看，随机森林 AUC 曲线在树数量增加后逐渐稳定，说明模型没有明显依赖单棵树的偶然划分；历史害情模型的训练曲线也呈现稳定下降趋势，说明滞后特征能够为高风险识别提供有效信息。从性能评价看，结构化风险模型不只是给出一个分类结果，还能提供可解释的风险来源：若气象风险较高但视觉检测较弱，系统可提示加强巡查；若视觉检测置信度较高且历史风险持续上升，则应提高防治优先级。

因此，7.5 节的意义在于把视觉检测从“发现目标”扩展为“解释风险”。后续无人机作业规划将综合风险值、病害面积和检测置信度转化为任务优先级、喷洒药量和补给策略，从而形成从识别到决策的闭环。


### 7.6 无人机路径规划与变量施药仿真实验

完成视觉识别和风险预测后，系统还需要回答“如何执行防治”。如果只给出风险等级，而不能转化为无人机任务、喷洒路径和药量分配，模型结果就难以服务实际植保作业。因此，本节将综合风险结果进一步映射为空间任务点，并在资源约束下完成多无人机路径规划和变量施药仿真。

本节按照实际作业流程展开：先构建数字孪生农田底图，再生成病害风险区域和任务点，随后完成多无人机任务分配、斑块内部覆盖喷洒、药量估计、资源匮乏识别和道路补给点选择。这样可以验证本文方法是否能够从“风险判断”进一步走向“作业方案”。


#### 7.6.1 数字孪生农田底图构建

路径规划首先需要明确空间约束，包括农田边界、道路、水体、村庄和可补给区域等要素。数字孪生底图的作用是为病害任务点、无人机飞行路径和补给点选址提供统一坐标系，避免后续路径规划停留在抽象表格层面。

下方代码调用农田建模脚本生成一比一农田空间模型；若当前环境缺少部分地理几何依赖，则使用备用绘图逻辑生成可复现的示意底图，保证 notebook 能够稳定展示空间建模结果。

下方结果展示内部喷洒距离、预计药量、单位面积药量和沿程流量，用于说明综合风险如何转化为可执行的喷洒参数。

#### 7.6.5 资源消耗与道路补给点选址

## 九、结论与展望

### 9.1 研究结论

本文围绕农作物病虫害精准防控需求，构建了面向植保无人机影像的多源数据融合智能风险检测系统。视觉识别层面，本文以 YOLO11s 为基准模型，针对农田图像中病斑尺度小、边界模糊、背景干扰强和类别相似等问题，引入 MSAFE 多尺度注意力增强结构，使模型能够更充分地利用浅层细节和深层语义信息。实验结果表明，MSAFE 改进模型在 Recall、mAP@0.5 和 F1-score 上较 Baseline 有稳定提升，说明该结构对小目标病害检出和复杂背景鲁棒性具有实际作用。

风险建模层面，本文将气象因子、历史害情和视觉识别结果统一纳入综合风险评估框架。随机森林模型用于刻画当前气象环境对病虫害发生的影响，XGBoost/GradientBoosting 模型用于建模历史害情延续趋势，视觉风险则由检测置信度和病害面积共同构成。三类风险共同决定地块防治优先级，弥补了单一图像识别无法反映环境诱因和历史趋势的不足。

无人机作业层面，本文进一步将综合风险转化为任务优先级、病害斑块、覆盖喷洒路径、变量施药量和道路补给点选择，实现了从“看见病害”到“判断风险”再到“指导施药”的闭环。平台实现部分验证了该框架具有一定工程集成能力，可为农田病虫害数字化监测和精准植保提供参考。

### 9.2 不足与展望

本文仍存在一些不足。首先，图像数据虽然覆盖水稻和玉米多类病虫害，但不同类别之间仍存在样本不均衡，少样本类别和小虫体目标的泛化能力仍需进一步验证。其次，风险模型目前主要基于结构化气象和历史害情数据，尚未充分引入遥感时序、土壤墒情和作物生育期等信息。再次，无人机作业规划仍以仿真环境为主，实际部署中还需要考虑风场、障碍物、飞控误差和药液漂移等复杂因素。

后续工作可从三个方向继续改进：第一，扩展多地区、多季节、多作物病虫害数据，提升模型跨区域泛化能力；第二，引入轻量化部署和边缘推理优化，使模型更适合无人机端实时运行；第三，将真实飞行轨迹、农药喷洒反馈和防治效果评价纳入闭环学习，使系统从“离线决策”进一步走向“在线自适应优化”。
